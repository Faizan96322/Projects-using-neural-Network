```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

```python
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
dataframe = pd.DataFrame(breast_cancer_dataset.data,columns = breast_cancer_dataset.feature_names)
dataframe.head()
```

```python
dataframe['label'] = breast_cancer_dataset.target
dataframe.tail()
```

```python
dataframe.shape
dataframe.info()
```

```python
X = dataframe.drop(columns='label', axis=1)
Y = dataframe['label']
print(X)
```

```python
print(Y)
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size = 0.2, random_state=2)
X_train.shape, X.shape, X_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
print(X_train)
```

```python
import tensorflow as tf
tf.random.set_seed(3)
ann = tf.keras.models.Sequential()
```

```python
ann.add(tf.keras.layers.Flatten(input_shape=(30,)))
ann.add(tf.keras.layers.Dense(units=30,activation='relu'))
ann.add(tf.keras.layers.Dense(units=20,activation='relu'))
ann.add(tf.keras.layers.Dense(units=2,activation='sigmoid'))

#compiling:

ann.compile(optimizer = 'adam', loss = 'sparse_categorical_crossentropy', metrics = ['accuracy'])
```

```python
#training:

ann.fit(X_train,Y_train, validation_split=0.1, epochs = 100)
```

```python
accuracy = ann.evaluate(X_test, Y_test)

print(accuracy)
```
